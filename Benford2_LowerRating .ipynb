```python
import numpy as np
import pandas as pd 
import benford as bf
```

```python
sp =pd.read_csv('train1.csv',index_col='handle')
##Taken 3000 data points for checking 
```

```python
sp['p_r'] =sp.old_rating/sp.old_rating.shift()-1 #simple returns 
```

```python
sp['l_r']=np.log(sp.old_rating/sp.old_rating.shift()) #log returns 
```

```python
sp.head()
```

```python
sp.tail()
```

```python
f1d = bf.first_digits(sp.l_r, digs=1, decimals=4) # digs=1 for the first digit (1-9)
```

```python
f2d = bf.first_digits(sp.l_r, digs=2, decimals=8) # digs=2!
```

```python

# For a significance of 5% (p <= 0.05), a confidence of 95%
f2d = bf.first_digits(sp.l_r, digs=2, decimals=8, confidence=95)
```

```python
f3d=bf.first_digits(sp.l_r,digs=3,decimals=8)
```

```python
## Taking the zscores of the first 3 digits ,with 85% confidence level 
f3d=bf.first_digits(sp.l_r,digs=3,decimals=8,confidence=85)
```

```python
## Last 2 digits test with confidence 85% and showing the negative deviations 
l2d= bf.last_two_digits(sp.l_r,decimals=8,confidence=85,high_Z='neg')
```

```python
## Mantissas 
mant1 = bf.mantissas(sp.l_r)
```

```python
old_rating_mant = bf.mantissas(sp.old_rating,report=True,show_plot=True)
```

```python
## The benford class report for precison 1 
benf= bf.Benford((sp,'l_r'),decimals=8)
benf.F1D.report()
```

```python
## Benford class report for 2d precision 
benf.F2D.report()
```

```python
##Benford 3 degree precision 
benf.F3D.report()
```

```python
##benford sd report
benf.SD.report()
```
